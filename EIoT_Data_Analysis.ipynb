```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(style="whitegrid")
```

```python
dataset = pd.read_csv(("SensorData.csv"))
dataset.head(10)
```

```python
dataset.columns
```

```python
data = dataset.drop(['rain_level', 'acceleration_x', 'acceleration_y', 'acceleration_z', 'inclination_xz', 'inclination_yz', 'max_wind_velocity', 'direct_max_wind_velocity', 'max_inst_wind_velocity'], axis=1)
```

```python
sns.histplot(data=data, x = "illuminance", kde=True)
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(data.corr(),annot=True,cmap='terrain')
```

```python
sns.violinplot(y="temperature", x="illuminance", data=data)
```

```python
data.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```
